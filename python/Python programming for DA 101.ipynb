````python
## control flow: if, for, while,
def grading(score):
    """
    input: score is a numberic number
    output: grade passed or failed
    """
    if score > 80:
        return "Passed"
    else:
        return "Failed"
grading(81)
````

````python
## multiple if else
def full_grading(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"
full_grading(80)
````

````python
## if multiple conditions
# morning weekday => cereal
# morning weekend => hamburger
# else => fasting

time = "no"
day = "weekend"

if time == "morning" and day == "weekday":
    print("I'm eating cereal")
elif time == "morning" and day == "weekend":
    print("I'm eating hamburger")
else:
    print("I'm eating nothing, I'm fasting")
````
